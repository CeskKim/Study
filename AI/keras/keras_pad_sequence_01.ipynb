```
from keras.preprocessing import sequence
import numpy as np
```

```
List = np.array([[2,4,6], [3,7,8,9], [5,10]])
display(List)
```

```
ListPad = sequence.pad_sequences(List) #pad_sequences를 이용, 문자길이 동일
display(ListPad)
```

```
ListPad = sequence.pad_sequences(List ,padding='pre', value=1 ) #pad_sequences를 이용, value1로 지정, pre를 통해 시퀀스 앞쪽으로 패딩
display(ListPad)
```

```
ListPad = sequence.pad_sequences(List ,padding='post', value=1 ) #pad_sequences를 이용, value1로 지정, post를 통해 시퀀스 뒤쪽으로 패딩
display(ListPad)
```
